# **Feature Extraction of Text data using Tf-idf Vectoizer**

Feature Extraction --> Converting Text data into Numerical data

About the dataset:

1. id: unique id for a news article
2. title: the title of a news article
3. author: author of the news article
4. text: the text of the article; could be incomplete
5. label: a label that marks whether the news article is real or fake: 1: Fake news 0: real News

Importing the Dependencies

```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
nltk.download('stopwords')
```

```python
# printing the stopwords in English
print(stopwords.words('english'))
```

```python
data = pd.read_csv('/content/fake_news_dataset.csv', on_bad_lines='skip')
```

```python
data.head()
```

```python
data.shape
```

```python
# counting the number of missing values in the dataset
data.isnull().sum()
```

```python
# replacing the null values with empty string
data = data.fillna('')
```

```python
# merging the author name and news title
data['content'] = data['author']+' '+data['title']
```

```python
print(data['content'])
```

```python
# separating the data & label
X = data.drop(columns='label', axis=1)
Y = data['label']
```

```python
print(X)
print(Y)
```

Stemming:

Stemming is the process of reducing a word to its Root word

example: actor, actress, acting --> act

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content
                       if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
data['content'] = data['content'].apply(stemming)
```

```python
print(data['content'])
```

```python
X = data['content'].values
Y = data['label'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
Y.shape
```

Tf-idf

```python
# convert the textual data to Feature Vectors
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit(X)
X = vectorizer.transform(X)
```
