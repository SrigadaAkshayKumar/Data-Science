```python
# creating an array of Zeros
x=np.zeros((3,3), dtype = int)
print(x)
```

```python
# creating an array of Ones
x=np.ones((3,3), dtype = int)
print(x)
```

```python
# creating an array with a specific value
x=np.full((3,3),4)
print(x)
```

```python
# printing identity matrix
x=np.eye(3)
print(x)
```

```python
# matrix with random values
x=np.random.randint(10,100,(3,2))
print(x)
```

Array Manipulation

```python
trans = np.transpose(x)
print(trans)
```
